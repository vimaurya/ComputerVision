```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image_path = '../data/goku.jpg'

image = cv.imread(image_path)

image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
kernel = np.ones((3,3), np.uint8)
```

```python
opened_image = cv.morphologyEx(image, cv.MORPH_OPEN, kernel)
closed_image = cv.morphologyEx(image, cv.MORPH_CLOSE, kernel)
gradient_image = cv.morphologyEx(image, cv.MORPH_GRADIENT, kernel)
tophat_image = cv.morphologyEx(image, cv.MORPH_TOPHAT, kernel)
blackhat_image = cv.morphologyEx(image, cv.MORPH_BLACKHAT, kernel)
#
# cv.imshow('original', image)
# cv.imshow('open', opened_image)
# cv.imshow('close', closed_image)
# cv.imshow('gradient', gradient_image)
# cv.imshow('tophat', tophat_image)
# cv.imshow('blackhat', blackhat_image)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
titles = ['Original Image', 'Opened Image', 'Closed Image', 'Gradient Image', 'Top Hat Image', 'Black Hat Image']
images = [image, opened_image, closed_image, gradient_image, tophat_image, blackhat_image]

plt.figure(figsize=(15, 10))
for i in range(6):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
