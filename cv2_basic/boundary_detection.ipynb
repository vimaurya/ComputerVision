```python
import cv2 as cv
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
img = cv.imread(os.path.join('..','data','kobe.jpg'))

img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python

kernel = np.array([[1,1,1],
                   [1,1,1],
                   [1,1,1]], dtype=np.uint8)

img_edge_erode = cv.erode(img, kernel, iterations=1)

diff = cv.absdiff(img, img_edge_erode)


# cv.imshow('edge', diff)
# cv.waitKey(0)
# cv.destroyAllWindows()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(img)
axes[0].set_title('original')
axes[0].axis('off')

axes[1].imshow(diff)
axes[1].set_title('boundaries')
axes[1].axis('off')

plt.show()
```

```python
lines = cv.imread(os.path.join('..','data','lines.jpg'))

lines = cv.cvtColor(lines, cv.COLOR_BGR2GRAY)

kernel1 = np.ones((1,8), dtype = np.uint8)

lines_erode = cv.erode(lines, kernel1, iterations=1)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(lines)
axes[0].set_title('lines')
axes[0].axis('off')

axes[1].imshow(lines_erode)
axes[1].set_title('horizontal lines')
axes[1].axis('off')

plt.show()
```
