```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('../data/bird.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

blur_kernel = np.ones((5, 5), np.float32) / 25

sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])

# Edge detection kernels
sobel_kernel = np.array([[-1, 0, 1],
                         [-2, 0, 2],
                         [-1, 0, 1]])

scharr_vertical = np.array([[ 3, 10,  3],
                            [ 0,  0,  0],
                            [-3,-10, -3]])

blurred = cv.filter2D(image, -1, blur_kernel)
sharpened = cv.filter2D(image, -1, sharpen_kernel)
edges = cv.filter2D(image, cv.CV_64F, sobel_kernel)
edges_scharr = cv.filter2D(image, -1, scharr_vertical)

# cv.imshow('Original ', image)
# cv.imshow('Blurred', blurred)
# cv.imshow('Sharpened', sharpened)
# cv.imshow("Edges", edges)
# cv.waitKey(0)
# cv.destroyAllWindows()

fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes[0].imshow(image)
axes[0].set_title("Original")
axes[0].axis('off')

axes[1].imshow(blurred)
axes[1].set_title("Blurred")
axes[1].axis('off')

axes[2].imshow(sharpened)
axes[2].set_title("Sharpened")
axes[2].axis('off')

axes[3].imshow(edges_scharr)
axes[3].set_title("Edges")
axes[3].axis('off')

plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('../data/kobe.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)



kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

laplacian_kernel = np.array([[-1, -1, -1],
                            [-1,  8, -1],
                            [-1, -1, -1]])


filter = cv.filter2D(image, -1, laplacian_kernel)


# cv.imshow('filter',filter)
# cv.waitKey(0)
# cv.destroyAllWindows()


fig, ax = plt.subplots(figsize=(6, 5))

ax.imshow(filter)
ax.set_title("filter")
ax.axis('off')
```
